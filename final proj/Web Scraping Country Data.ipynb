### WebScraping Process to Find Covid-19 Data by Country

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url = 'https://www.worldometers.info/coronavirus/#countries'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
response
```

```python
table_labels =  soup.find("table", attrs={"id": "main_table_countries_today"})
table_labels_data = table_labels.find_all("td")
```

```python
# filtering the data that I will going to be using in my table
extract = table_labels_data[121:3329]
```

```python
#Creating the list of countries
extract = table_labels_data[121:3329][::15]
country_list = []
for x in extract:
    country_list.append(x.get_text().replace('\n', ''))
```

```python
#Total Cases list
extract = table_labels_data[122:3329][::15]
total_cases = []
for x in extract:
    total_cases.append(x.get_text().replace('\n', '').replace(',', ''))

    
total_cases_copy = []
for x in total_cases:
    total_cases_copy.append(float(x))   
```

```python
#New Cases list
extract = table_labels_data[123:3329][::15]
new_cases = []
for x in extract:
    new_cases.append(x.get_text().replace('\n', '').replace('+', '').replace(',', ''))
    
new_cases_copy = []
for x in new_cases:
    if x == '':
        x = None
        new_cases_copy.append(x)
    else:
        new_cases_copy.append(float(x))
```

```python
#Total Deaths list
extract = table_labels_data[124:3329][::15]
total_deaths = []
for x in extract:
    total_deaths.append(x.get_text().replace('\n', '').replace(',', ''))
    
    
total_deaths_copy = []
for x in total_deaths:
    if x == ' ':
        x = None
        total_deaths_copy.append(x)
    else:
        total_deaths_copy.append(float(x))
```

```python
#New Deaths list
extract = table_labels_data[125:3329][::15]
new_deaths = []
for x in extract:
    new_deaths.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))
    
    
new_deaths_copy = []
for x in new_deaths:
    if x == '':
        x = None
        new_deaths_copy.append(x)
    else:
        new_deaths_copy.append(float(x))
```

```python
#Totaled Recovered list
extract = table_labels_data[126:3329][::15]
total_recovered = []
for x in extract:
    total_recovered.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))

    
total_recovered_copy = []
for x in total_recovered:
    if x == '' or  x == "N/A":
        x = None
        total_recovered_copy.append(x)
    else:
        total_recovered_copy.append(float(x))
```

```python
#Active Cases list
extract = table_labels_data[127:3329][::15]
active_cases = []
for x in extract:
    active_cases.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))

    
active_cases_copy = []
for x in active_cases:
    if x == '' or  x == "N/A":
        x = None
        active_cases_copy.append(x)
    else:
        active_cases_copy.append(float(x))
```

```python
#Serious Critical list
extract = table_labels_data[128:3329][::15]
serious_critical = []
for x in extract:
    serious_critical.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))

    
serious_critical_copy = []
for x in serious_critical:
    if x == '':
        x = None
        serious_critical_copy.append(x)
    else:
        serious_critical_copy.append(float(x))
```

```python
#Total Cases/1M Pop list
extract = table_labels_data[129:3329][::15]
total_cases_per_mil = []
for x in extract:
    total_cases_per_mil.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))


total_cases_per_mil_copy = []
for x in total_cases_per_mil:
    if x == '':
        x = None
        total_cases_per_mil_copy.append(x)
    else:
        total_cases_per_mil_copy.append(float(x))
```

```python
#Deaths/1M Pop list
extract = table_labels_data[130:3329][::15]
deaths_per_mil = []
for x in extract:
    deaths_per_mil.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))


deaths_per_mil_copy = []
for x in deaths_per_mil:
    if x == '':
        x = None
        deaths_per_mil_copy.append(x)
    else:
        deaths_per_mil_copy.append(float(x))
```

```python
#Total Tests list
extract = table_labels_data[131:3329][::15]
total_tests = []
for x in extract:
    total_tests.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))


total_tests_copy = []
for x in total_tests:
    if x == '':
        x = None
        total_tests_copy.append(x)
    else:
        total_tests_copy.append(float(x))
```

```python
#Tests/1M Pop list
extract = table_labels_data[132:3329][::15]
tests_per_mil = []
for x in extract:
    tests_per_mil.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))

    
tests_per_mil_copy = []
for x in tests_per_mil:
    if x == '':
        x = None
        tests_per_mil_copy.append(x)
    else:
        tests_per_mil_copy.append(float(x))
```

```python
#Population list
extract = table_labels_data[133:3329][::15]
population = []
for x in extract:
    population.append(x.get_text().replace('\n', '').replace(',', '').replace('+', ''))

    
population_copy = []
for x in population:
    if x == " ":
        x = None
        population_copy.append(x)
    else:
        population_copy.append(float(x))
```

### Final Data 

```python
dataframe_data = {'Total Cases': total_cases_copy , 'New Cases': new_cases_copy, 
                  'Total Deaths':total_deaths_copy, 'New Deaths': new_deaths_copy, 
                  'Total Recovered': total_recovered_copy, 'Active Cases': active_cases_copy, 
                  'Serious Critical': serious_critical_copy, 'Total Cases/ 1M Pop': total_cases_per_mil_copy, 
                  'Death/ 1M Pop': deaths_per_mil_copy, 'Total Tests': total_tests_copy, 
                  'Tests/ 1M Pop': tests_per_mil_copy, 'Population': population_copy}
row_labels = country_list
```

```python
import pandas
df = pandas.DataFrame(data=dataframe_data, index=row_labels)
len(df)
```
